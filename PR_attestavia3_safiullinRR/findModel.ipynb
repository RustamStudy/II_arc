```python
import torch
from transformers import AutoTokenizer, AutoModel, AutoModelForQuestionAnswering
from transformers import FSMTForConditionalGeneration, FSMTTokenizer
from transformers import BertForQuestionAnswering
from transformers import BertTokenizer, BertModel

import sacremoses
import numpy as np
```

```python
question = "люди были на луне?"
paragraph = "В период с 1969 по 1972 год по программе «Аполлон» было выполнено 6 полётов с посадкой на Луне. Всего на Луне высаживались 12 астронавтов США. Список космонавтов Список космонавтов — участников орбитальных космических полётов Список астронавтов США — участников орбитальных космических полётов Список космонавтов СССР и России — участников космических полётов Список женщин-космонавтов Список космонавтов, посещавших МКС Энциклопедия астронавтики"
```

```python
modelNameRuEn = "facebook/wmt19-ru-en"

tokenizerRuEn = FSMTTokenizer.from_pretrained(modelNameRuEn)
modelRuEn = FSMTForConditionalGeneration.from_pretrained(modelNameRuEn)
```

```python
modelNameEnRu = "facebook/wmt19-en-ru"

tokenizerEnRu = FSMTTokenizer.from_pretrained(modelNameEnRu)
modelEnRu = FSMTForConditionalGeneration.from_pretrained(modelNameEnRu)
```

```python
def translateRuEn(inputRu):
  input = inputRu
  input_ids = tokenizerRuEn.encode(input, return_tensors="pt")
  outputs = modelRuEn.generate(input_ids)
  decodedEn = tokenizerRuEn.decode(outputs[0], skip_special_tokens=True)
  return decodedEn
```

```python
def translateEnRu(inputEn):
  input = inputEn
  input_ids = tokenizerEnRu.encode(input, return_tensors="pt")
  outputs = modelEnRu.generate(input_ids)
  decodedRu = tokenizerEnRu.decode(outputs[0], skip_special_tokens=True)
  return decodedRu
```

```python
print(paragraph)
```

```python
print(translateRuEn(paragraph))
```

```python
paragraph_ru_en_ru = translateRuEn(paragraph)
print(paragraph_ru_en_ru)
```

Расмотрим русские и английские библиотеки обработки

```python
questionEn = translateRuEn(question)
paragraphEn = translateRuEn(paragraph)
```

```python
modelNameQa1 = "bert-large-uncased-whole-word-masking-finetuned-squad"
modelNameQa2 = "distilbert-base-cased-distilled-squad"
modelNameQa3 = "distilbert-base-uncased-distilled-squad"
modelNameQa4 = "mrm8488/bert-medium-finetuned-squadv2"
modelNameQa5 = "bert-large-cased-whole-word-masking-finetuned-squad"
modelNameQa6 = "AlexKay/xlm-roberta-large-qa-multilingual-finedtuned-ru"
```

```python
tokenizer_1 = BertTokenizer.from_pretrained(modelNameQa1)
model_1 = BertForQuestionAnswering.from_pretrained(modelNameQa1)

input_ids_1 = tokenizer_1.encode(questionEn, paragraphEn)
```

```python
tokenizer_2 = AutoTokenizer.from_pretrained(modelNameQa2)
model_2 = AutoModel.from_pretrained(modelNameQa2)

input_ids_2 = tokenizer_2.encode(questionEn, paragraphEn)
```

```python
tokenizer_3 = AutoTokenizer.from_pretrained(modelNameQa3)
model_3 = AutoModel.from_pretrained(modelNameQa3)

input_ids_3 = tokenizer_3.encode(questionEn, paragraphEn)
```

```python
tokenizer_4 = AutoTokenizer.from_pretrained(modelNameQa4)
model_4 = AutoModel.from_pretrained(modelNameQa4)

input_ids_4 = tokenizer_4.encode(questionEn, paragraphEn)
```

```python
tokenizer_5 = AutoTokenizer.from_pretrained(modelNameQa5)
model_5 = AutoModel.from_pretrained(modelNameQa5)

input_ids_5 = tokenizer_5.encode(questionEn, paragraphEn)
```

```python
tokenizer_6 = AutoTokenizer.from_pretrained(modelNameQa6)
model_6 = AutoModelForQuestionAnswering.from_pretrained(modelNameQa6)

input_ids_6 = tokenizer_6.encode(question, paragraph)
```

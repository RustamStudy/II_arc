```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
```

```python
MODEL_FILE_EFF = './people_university_model_eff.cbm'
MODEL_FILE = './people_university_model.cbm'
```

Получение датасета для дальнейшей работы


```python
df = pd.read_csv('./student.csv')
df
```

```python
X = df.drop(columns=['final_lab', 'mean_score'], axis=1)
X
```

```python
y = df['final_lab'] #Series
y
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
print(y)
```

```python
def trainTestSplit(X, y, test_size, random_state):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    return [X_train, X_test, y_train, y_test]
```

```python
def CatBoostClassifier__def(X,y,test_size,iterations,depth,learning_rate,modelFile):
  random_state = 1234
  trainTestSplitArr = trainTestSplit(X, y, test_size, random_state)
  X_train = trainTestSplitArr[0]
  X_test = trainTestSplitArr[1]
  y_train = trainTestSplitArr[2]
  y_test = trainTestSplitArr[3]

  model = CatBoostClassifier(iterations=iterations, depth=depth, learning_rate=learning_rate)
  model.fit(X_train, y_train)
  model.save_model(modelFile)

  y_pred = model.predict(X_test)

  return [y_test, y_pred]
```

```python
def model_to_work(X,y,modelFile):
  y_Arr = CatBoostClassifier__def(X,y,test_size=0.2,iterations=100, depth=4, learning_rate=0.1,modelFile = modelFile)

  y_test = y_Arr[0]
  y_pred = y_Arr[1]

  accuracy = accuracy_score(y_test, y_pred)
  print(f"Правильность (accuracy) модели: {accuracy}")

  precision = precision_score(y_test, y_pred, average='weighted')
  print(f"Точность (precision) модели: {precision}")

  recall = recall_score(y_test, y_pred, average='weighted')
  print(f"Полнота (recall) модели: {recall}")

  f1 = f1_score(y_test, y_pred, average='weighted')
  print(f"F1 мера модели: {f1}")
```

```python
model_to_work(X,y,MODEL_FILE)
```

```python
selector = SelectKBest(chi2, k=3)
X_new = selector.fit_transform(X, y)

model_to_work(X_new,y,MODEL_FILE_EFF)
```

```python
def classifyStudent(new_student, model_file):
    model = CatBoostClassifier()
    model.load_model(model_file)
    new_data = pd.DataFrame([new_student])
    predicted_category = model.predict(new_data)
    return predicted_category[0][0]
```

```python
# число предметов
SUBJ_COUNT = 6
# мин балл
MIN_GRADE = 0
# макс балл
MAX_GRADE = 99

new_grades = np.random.randint(MIN_GRADE, MAX_GRADE + 1, SUBJ_COUNT)
new_grades
```

```python
result = classifyStudent(new_grades, MODEL_FILE)

print(f"Обычная модель - {result}")

result = classifyStudent(new_grades, MODEL_FILE_EFF)

print(f"Модель с SelectBestK: {result}")
```

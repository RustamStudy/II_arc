Проверка на корректность матрицы


```python
def matrixValidation(A):
 if(len(A) > 0 and hasattr(A[0], "__len__") and len(A[0]) > 0):
  for i in range(len(A)):
    if(len(A[i])!=len(A[0])):
     return False
 else:
  return False

 return True
```

Проверка на корректность матриц для произведения

```python
def multiplyMatrixValidation(A, B):
   if(matrixValidation(A) == False):
    return 'А - не матрица'
   if(matrixValidation(B) == False):
    return 'B - не матрица'
   if(len(A[0]) != len(B)):
    return 'Произведение неопределено число столбцов первой матрицы не совпадает с числом строк второй матрицы'
   return True
```

Перемножение матриц

```python
def matrixMultiply(A, B):
  resultMultiplyMatrixValidation = multiplyMatrixValidation(A, B)
  if(resultMultiplyMatrixValidation == True):
    C = [[0 for i in range(len(B[0]))] for i in range(len(A))]
    for i in range(len(A)):
      for j in range(len(A[0])):
        for k in range(len(B[0])):
          C[i][k] += A[i][j] * B[j][k]
    return C
  else:
    return resultMultiplyMatrixValidation
```

```python
A1 = [
      [6, -8, 3]
     ]
B1 = [
      [1],
      [1],
      [4]
     ]
matrixMultiply(A1, B1)
```

```python
A2 = [
      [-3, -7],
      [-10, 6],
      [0, -6],
      [-1, -3]
     ]
B2 = [
      [-8, 9],
      [4, -2]
     ]
matrixMultiply(A2, B2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataSet= pd.read_csv('Titanic-Dataset.csv', delimiter=',')
dataSet
```

# Выжившие и не выжившие по отношению к количеству братьев, сестер....

```python
dataSet_SibSp_Unigue = dataSet['SibSp'].sort_values().unique()

dataSet_SibSp = []
for i in range(len(dataSet_SibSp_Unigue)):
  dataSet_SibSp.append([0,0])
  for j in range(len(dataSet)):
    if dataSet['SibSp'][j] == dataSet_SibSp_Unigue[i]:
      if dataSet['Survived'][j] == 1:
        dataSet_SibSp[i][0] = dataSet_SibSp[i][0] + 1
      if dataSet['Survived'][j] == 0:
        dataSet_SibSp[i][1] = dataSet_SibSp[i][1] + 1


dataSet_SibSp = pd.DataFrame(dataSet_SibSp, dataSet_SibSp_Unigue, ['Выжил','Не выжил'])

dataSet_SibSp
```

```python
dataSet_SibSp.plot.bar(title = 'Выжившие и не выжившие по отношению к количеству братьев, сестер....')
```

```python
dataSet_SibSp['%Survived'] = dataSet_SibSp['Выжил'] * 100/(dataSet_SibSp['Выжил'] + dataSet_SibSp['Не выжил'])
dataSet_SibSp['%Survived'].plot.bar(title = 'Выжившие и не выжившие по отношению к количеству братьев, сестер.... - В процентах')
```

# Выживаемость в зависимости от пола и возраста

```python
index = []
setAge = []
for i in range(int(dataSet['Age'].max()/10)):
  index.append('['+ str(i * 10) + '-' + str((i + 1) * 10) + ')')
  setAge.append(['male',0,0,'female',0,0,0,0])
  for j in range(len(dataSet)):
    if dataSet['Age'][j] >= i * 10 and dataSet['Age'][j] < (i + 1) * 10:
      if dataSet['Sex'][j] == 'male':
        if dataSet['Survived'][j] == 1:
          setAge[i][1] = setAge[i][1] + 1
        if dataSet['Survived'][j] == 0:
          setAge[i][2] = setAge[i][2] + 1
      if dataSet['Sex'][j] == 'female':
        if dataSet['Survived'][j] == 1:
          setAge[i ][4] = setAge[i][4] + 1
        if dataSet['Survived'][j] == 0:
          setAge[i][5] = setAge[i][5] + 1
  setAge[i][6] = setAge[i][1] + setAge[i][4]
  setAge[i][7] = setAge[i][2] + setAge[i][5]
setAge = pd.DataFrame(setAge, index, ['Пол1','Выжил_М','Не выжил_М','Пол2','Выжил_Ж','Не выжил_Ж','Итого_Выжил', 'Итого_Не_выжил'])
setAge
```

```python
setAge.plot.bar(y=['Итого_Выжил','Итого_Не_выжил'], title = 'Распределение по возрастам')
```

```python
setAge['%Survived_Итог_Выжил'] = setAge['Итого_Выжил'] * 100/(setAge['Итого_Выжил'] + setAge['Итого_Не_выжил'])
setAge['%Survived_Итог_Не_выжил'] = setAge['Итого_Не_выжил'] * 100/(setAge['Итого_Выжил'] + setAge['Итого_Не_выжил'])
setAge.plot.bar(y=['%Survived_Итог_Выжил','%Survived_Итог_Не_выжил'],title = 'Распределение по возрастам - В процентах')
```

```python
setAge.plot.bar(y=['Выжил_М','Не выжил_М'], title = 'Распределение по возрастам (Мужчины)')
```

```python
setAge['%Survived_Итог_Выжил_М'] = setAge['Выжил_М'] * 100/(setAge['Выжил_М'] + setAge['Не выжил_М'])
setAge['%Survived_Итог_Не_выжил_М'] = setAge['Не выжил_М'] * 100/(setAge['Выжил_М'] + setAge['Не выжил_М'])
setAge.plot.bar(y=['%Survived_Итог_Выжил_М','%Survived_Итог_Не_выжил_М'],title = 'Распределение по возрастам (Мужчины) - В процентах')
```

```python
import pandas as pd
import os

from datetime import datetime,timedelta
```

```python
def getFileName(dir):
    directory = '../../'+dir+'/'
    fileArray = os.listdir(directory)
    return fileArray
```

```python
dir = 'climate'
dfArray = pd.DataFrame()
fileList = getFileName(dir)
for i in range(len(fileList)):
    df = pd.read_csv('../../'+dir+'/'+fileList[i], delimiter=';').dropna()
   
    df['day_begin'] = df['day'].apply(
        lambda x: datetime.strptime(x,'%d.%m.%Y %H:%M'))
    df['day_end'] = df['day_begin'] + timedelta(hours=1)
    dfArray.append(df)
    if i == 0:
        dfArray = df
    else:
        dfArray = dfArray.append(df)
dfArray
```

```python
dfArray.to_csv('../../readyData/climatRoom.csv')
```

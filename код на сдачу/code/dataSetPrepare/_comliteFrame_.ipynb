```python
import pandas as pd
import numpy as np
import os

from datetime import datetime,timedelta
```

```python
dfParameter = pd.read_csv('../../readyData/newFrame.csv')
dfParameter
```

```python
resultArray = pd.DataFrame(dfParameter['result'].unique(), columns=['result_name'])
resultArray
```

```python
resultArray.to_csv('../../readyData/resultList.csv', index=False)
```

```python
  
dfParameter['result_ind'] = ''
for i in range(len(dfParameter)):
    dfParameter['result_ind'].iloc[i] = np.int_(resultArray[resultArray['result_name'] == dfParameter['result'].iloc[i]].index.values)[0]
dfParameter
```

```python
dfPress = pd.read_csv('../../readyData/pressParamrter_.csv')
dfPress
```

```python
dfParameter['press'] = ''
dfParameter['T_steam'] = ''
for i in range(len(dfParameter)):
    dfParameter['press'].iloc[i] = np.float64(dfPress['Pres'].loc[((dfPress['Number'] == dfParameter['id_marsh'].iloc[i]))])
    dfParameter['T_steam'].iloc[i] = np.float64(dfPress['Temp'].loc[((dfPress['Number'] == dfParameter['id_marsh'].iloc[i]))])
dfParameter
```

```python
dfTemp = pd.read_csv('../../readyData/climatRoom.csv')
dfTemp = dfTemp.drop(columns=['Unnamed: 0'], axis=1)
dfTemp
```

```python
dfTemp['day_begin_hour'] = dfTemp['day_begin'].apply(
    lambda x: x.split(':')[0]
)    
dfTemp
```

```python
dfParameter['day_begin_hour'] = dfParameter['date_in'].apply(
    lambda x: x.split(':')[0]
)    
dfParameter
```

```python
dfParameter['temp'] = ''
dfParameter['hydr'] = ''
for i in range(len(dfParameter)):
    dfParameter['temp'].iloc[i] = np.float64(dfTemp['temp'].loc[((dfTemp['day_begin_hour'] == dfParameter['day_begin_hour'].iloc[i]))])
    dfParameter['hydr'].iloc[i] = np.float64(dfTemp['hydr'].loc[((dfTemp['day_begin_hour'] == dfParameter['day_begin_hour'].iloc[i]))])
dfParameter
```

```python
totalFrame = dfParameter.drop(columns=['id_marsh','date_in','art_num','result','day_begin_hour'], axis=1)
totalFrame
```

```python
totalFrame.to_csv('../../readyData/totalDS.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_decision_regions

from sklearn.linear_model import LinearRegression

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
from datetime import datetime



from sklearn.model_selection import cross_val_score, KFold, train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression, RidgeClassifier, Lasso,SGDClassifier, BayesianRidge
from sklearn.naive_bayes import CategoricalNB, BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('./../readyData/totalDS.csv')
df
```

```python
X = df.drop(columns=['result_ind'], axis=1)
y = df['result_ind']
```

```python

folds = np.array([5,10,15,20,25])
# folds = 5
test= np.array([20,25,30])
metric = 'accuracy'
models = {}
```

```python
nameMethod = ['LogisticRegression','DecisionTreeClassifier','SVC','KNeighborsClassifier','RandomForestClassifier','BernoulliNB']

# Логическая регрессия
models['LogisticRegression'] = LogisticRegression(multi_class = 'multinomial')
# Naive Bayes Бернули
models['BernoulliNB'] = BernoulliNB()
# Дерево принятия решений
models['DecisionTreeClassifier'] = DecisionTreeClassifier()
# SVM
models['SVC'] = SVC()
# KNN
models['KNeighborsClassifier'] = KNeighborsClassifier(n_neighbors=10)
# RandomForest
models['RandomForestClassifier'] = RandomForestClassifier(n_estimators=150)
```

```python
def modelCalc(X,y,nameMethod,test):
    name = nameMethod
    print(name)
    print('start')
    print(datetime.now())
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test/100, random_state=1234)
    for i in range(len(folds)):
        print(f"folds - {folds[i]}")
        model = models[name]
        k_fold = KFold(n_splits=folds[i])
        res = cross_val_score(model,X_train,y_train,cv = k_fold,scoring=metric,n_jobs=-1)
        results.append([name,test,folds[i],res.mean()])
        print(f'value {res.mean()}')
    print('finish')
    print(datetime.now())
```

```python
results = []
for i in range(len(nameMethod)):
    for j in range(len(test)):
        print(f"метод - {nameMethod[i]}")
        print(f"Тестовая выборка - {test[j]}")
        
        modelCalc(X, y, nameMethod[i], test[j])
        
        print (f"\n")
```

```python
results
```

```python
resultsValue = pd.DataFrame(results)

resultsValue.loc[((resultsValue[3]==resultsValue[3].max()))]
```
